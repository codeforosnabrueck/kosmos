```python
import pandas as pd
```

```python
df = pd.read_csv('Excel_Export_Ziel3.csv')
df.head()
```

```python
df.iloc[0] = df.iloc[0].ffill()
df.head()
```

```python
df = df.iloc[:3, 1:]
df
```

```python
detailgoal = pd.Series(df.columns)
detailgoal[detailgoal.str.startswith('Unnamed')] = None
detailgoal = detailgoal.ffill()
detailgoal.tail()
```

```python
data = pd.DataFrame({
    'Detailziel': detailgoal,
    'Indikator': df.iloc[0].reset_index(drop=True),
    'Jahr': df.iloc[1].reset_index(drop=True),
    'Wert': df.iloc[2].reset_index(drop=True),
})
data
```

```python
data['Jahr'] = pd.to_datetime(data['Jahr'].astype(str).str[:4])
```

```python
try:
    data['Wert'] = pd.to_numeric(data['Wert'].astype(str))
except ValueError:
    data['Wert'] = data['Wert'].astype(str).str.replace(',', '')
    data['Wert'] = pd.to_numeric(data['Wert'])
```

```python
for group, subset in data.groupby('Indikator'):
    if len(subset) > 1:
        (ggplot(subset, aes(x='Jahr', y='Wert'))
         + geom_point(color='red')
         + geom_line(color='red')
         + geom_smooth(method='lm')
         + labs(title=group)
         + scale_x_date(date_breaks='1 year', date_labels='%Y')
        ).draw()
    else:
        continue
        
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for group, subset in data.groupby('Indikator'):
    if len(subset) > 1:
        subset['Jahr_int'] = subset.Jahr.dt.year
        sns.jointplot(x='Jahr_int', y='Wert', kind='reg', data=subset)
        plt.suptitle(group)
    else:
        continue
        
```

```python
data_year_indexed = data.set_index('Jahr')
data_year_indexed.head()
```

```python
subsets = []
for indicator, group in data_year_indexed.groupby('Indikator'):
    group = group.rename(columns={'Wert': indicator})
    group = group.drop(columns=['Indikator', 'Detailziel'])
    subsets.append(group)
data_corr = pd.concat(subsets, axis='columns')
data_corr
```

```python
data_corr.corr()
```

```python
corr = data_corr.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
