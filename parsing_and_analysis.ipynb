```python
import seaborn
#import altair as alt
from plotnine import *
```

```python
df = pd.read_csv('Excel_Export_Ziel3.csv')
df.head()
```

```python
df.iloc[0] = df.iloc[0].ffill()
df.head()
```

```python
df = df.iloc[:3, 1:]
df
```

```python
detailgoal = pd.Series(df.columns)
detailgoal[detailgoal.str.startswith('Unnamed')] = None
detailgoal = detailgoal.ffill()
detailgoal.tail()
```

```python
data = pd.DataFrame({
    'Detailziel': detailgoal,
    'Kennzahl': df.iloc[0].reset_index(drop=True),
    'Jahr': df.iloc[1].reset_index(drop=True),
    'Wert': df.iloc[2].reset_index(drop=True),
})
data
```

```python
data['Jahr'] = pd.to_datetime(data['Jahr'].astype(str).str[:4])
```

```python
try:
    data['Wert'] = pd.to_numeric(data['Wert'].astype(str))
except ValueError:
    data['Wert'] = data['Wert'].astype(str).str.replace(',', '')
    data['Wert'] = pd.to_numeric(data['Wert'])
```

```python
data
```

```python
data.info()
```

```python
for group, subset in data.groupby('Indikator'):
    if len(subset) > 1:
        (ggplot(subset, aes(x='Jahr', y='Wert'))
         + geom_point(color='red')
         + geom_line(color='red')
         + geom_smooth(method='lm')
         + labs(title=group)
         + scale_x_date(date_breaks='1 year', date_labels='%Y')
        ).draw()
    else:
        continue
        
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
for group, subset in data.groupby('Indikator'):
    if len(subset) > 1:
        subset['Jahr_int'] = subset.Jahr.dt.year
        sns.jointplot(x='Jahr_int', y='Wert', kind='reg', data=subset)
        plt.suptitle(group)
    else:
        continue
        
```

Turn each indicator into it's own column.

```python
data = data.pivot(index='Jahr', columns='Kennzahl', values='Wert')
data
```

```python
#dictionary that assigns the desired trend direction for each variable
trend_dict = {key: None for key in data.columns}
#all variables which we want to see rise over time:
trend_dict['Anteil Fußgänger an den Verkehrsmitteln in %'] = 'positive'
trend_dict['Anteil Radverkehr an den Verkehrsmitteln in %'] = 'positive'
trend_dict['Anteil zugelassener E-Fahrzeuge in %'] = 'positive'
trend_dict['Anteil zugelassener Plug-In Hybride am Gesamtbestand in %'] = 'positive'
trend_dict['Anteil ÖPNV an den Verkehrsmitteln in %'] = 'positive'
trend_dict['Anzahl Carsharing Fahrzeuge'] = 'positive'
trend_dict['Anzahl Carsharing Standorte'] = 'positive'
trend_dict['Anzahl der Personen, die sehr und eher zufrieden mit dem öffentlichem Nahverkehr sind'] = 'positive'
trend_dict['Beförderte Personen (ÖPNV)'] = 'positive'
trend_dict['Carsharing Nutzer'] = 'positive'

#all variables which we want to see diminish over time
trend_dict['Anteil PKW an den Verkehrsmitteln in %'] = 'negative'
trend_dict['Anzahl Fahrradunfälle  mit leichtverletzten Personen'] = 'negative'
trend_dict['Anzahl Fahrradunfälle  mit schwerverletzten Personen'] = 'negative'
trend_dict['Anzahl der Fahrradunfälle in der Stadt Osnabrück'] = 'negative'
trend_dict['Anzahl der Personen, die eher unzufrieden und überhaupt nicht zufrieden mit dem öffentlichem Nahverkehr sind'] = 'negative'
trend_dict['Anzahl der Verkehrsunfälle in der Stadt Osnabrück'] = 'negative'
trend_dict['Carsharing Nutzer'] = 'negative'
trend_dict['Carsharing Nutzer'] = 'negative'


trend_dict
```

```python
corr = data.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
